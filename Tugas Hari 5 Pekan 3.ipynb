```
def hexbin(x, y, color, **kwargs):

```


````python
#code here
def hexbin(x,y,color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x,y,gridsize=15, cmap=cmap, **kwargs)
    
g = sns.FacetGrid(df, col="cut")
g.map_dataframe(sns.scatterplot, x="carat", y="price")
g.set_axis_labels("carat", "price")
g.map_dataframe(hexbin)
````

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price sehingga dihasilkan output sesuai expected.

````python
df = df.sample(n=300, random_state=123)

#code here
sns.pairplot(df, hue="cut", vars = ['carat', 'depth', 'table','price']
            , diag_kind = 'kde')
````

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

````python
#code here
with sns.axes_style('white'):
    sns.jointplot("price", "carat", df, kind='kde')
    
with sns.axes_style('white'):
    sns.jointplot("price", "carat", df, kind='hex')
````
