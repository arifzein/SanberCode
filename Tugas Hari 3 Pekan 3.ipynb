```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
data1,data2,data3 = np.random.randn(3,12)

fig, ax = plt.subplots(nrows=1,ncols=3,sharex=True,sharey=True,figsize=(20,8))
l1, = ax[0].plot(x, tan, color='b')
l2, = ax[1].plot(x, sin, color='r')
l3, = ax[2].plot(x, cos, color='g')

ax[0].set_title('Grafik tanX')
ax[1].set_title('Grafik sinX')
ax[2].set_title('Grafik cosX')

fig.subplots_adjust(bottom=0.3, wspace=0.33)

ax[1].legend(handles = [l1,l2,l3] , labels=['tanX', 'sinX', 'cosX'],loc='upper center', 
             bbox_to_anchor=(0., 1.20, 1., .104),fancybox=True, shadow=False, ncol=1)
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
plt.figure(figsize=(12, 8))
ax = plt.axes()
ax2 = plt.axes((0.85, 0.65, 0.3, 0.3))
ax3 = plt.axes((-0.10, 0.65, 0.3, 0.3))
ax4 = plt.axes((-0.10, -0.00, 0.3, 0.3))
ax5 = plt.axes((0.85, 0.00, 0.3, 0.3))

ax.plot(x, y)
ax2.plot(x, y2, color='black')
ax2.plot(x, y, color='blue')
ax3.plot(x, y3, color='y')
ax3.plot(x, y, color='blue')
ax4.plot(x, y4, color='r')
ax4.plot(x, y, color='blue')
ax5.plot(x, y5, color='g')
ax5.plot(x, y, color='blue')

ax.set_title('Grafik CosX')
ax2.set_title('Grafik CosX^4')
ax3.set_title('Grafik CosX^5')
ax4.set_title('Grafik CosX^2')
ax5.set_title('Grafik CosX^3')
```
